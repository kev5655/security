```python
p = 11
prime_roots = [2,6,7,8]

x = range(1, 11)
y = range(1, 11)
```

## Aufgabe: Primitive Roots (Primitivwurzeln) modulo p = 11

**Ziel:** Finde ein Muster, das die primitiven Wurzeln {2, 6, 7, 8} identifiziert.

### Was ist eine Primitivwurzel (Primitive Root)?

Eine Zahl $g$ ist eine **Primitivwurzel modulo p**, wenn ihre Potenzen $g^1, g^2, g^3, \ldots, g^{p-1}$ **alle** Werte von 1 bis p-1 modulo p erzeugen (in irgendeiner Reihenfolge).

```python
import pandas as pd
import numpy as np

results = np.zeros((10, 10))

for x_idx, x_i in enumerate(x):
    for y_idx, y_i in enumerate(y):
        results[x_idx][y_idx] = pow(x_i, y_i, p)
        
df = pd.DataFrame(results, index=range(1, 11), columns=range(1, 11))
df.index.name = 'x (base)'
df.columns.name = 'y (exponent)'
print(df.to_string(float_format='%2.0f'))
```

```python
from pandas._typing import Scalar

def highlight_prime_roots(val: Scalar) -> str:
    """Highlight cells that show prime root behavior"""
    if val in prime_roots:
        return 'background-color: red'
    return ''

df.style.map(highlight_prime_roots).format('{:.0f}')
```

### Was zeigt die Tabelle?

Die Tabelle berechnet: $$x^y \bmod p$$

- **Zeilen (x)**: Die Basis (1 bis 10)
- **Spalten (y)**: Der Exponent (1 bis 10)
- Jede Zelle zeigt das Ergebnis von $x^y \bmod 11$

### Das Muster:

**Primitivwurzeln** (x = 2, 6, 7, 8) haben eine besondere Eigenschaft:
- In ihrer **Zeile** kommen **alle Werte von 1 bis 10** vor (keine Duplikate in den ersten p-1 Spalten)
- Sie "durchlaufen" alle möglichen Reste modulo 11 (außer 0)

**Nicht-Primitivwurzeln** (z.B. x = 3, 4, 5, 9):
- Erzeugen nur eine **Teilmenge** der Werte
- Haben **wiederholende Werte** in ihrer Zeile
- Erreichen nicht alle Zahlen von 1 bis 10

### Beispiel:
- x = 2 (Primitivwurzel): $2^1, 2^2, 2^3, \ldots$ erzeugt alle Werte 1-10
- x = 3 (keine Primitivwurzel): $3^1, 3^2, 3^3, \ldots$ erzeugt nur einige Werte mehrfach

```python
# Überprüfe jede Basis (Zeile): Erzeugt sie alle Werte von 1 bis p-1?
print(f"Überprüfung für p = {p}:")
print(f"Erwartete Primitivwurzeln: {prime_roots}\n")

for base in range(1, 11):
    # Hole alle Werte aus dieser Zeile (alle Potenzen von base)
    row_values = df.loc[base].values
    unique_values = set(row_values)
    
    # Eine Primitivwurzel erzeugt alle Werte von 1 bis p-1
    all_values = set(range(1, p))
    is_primitive = unique_values == all_values
    
    status = "✓ Primitivwurzel" if is_primitive else "✗ Keine Primitivwurzel"
    print(f"Basis {base}: {len(unique_values)} eindeutige Werte von {p-1} → {status}")
    print(f"  Erzeugte Werte: {sorted(unique_values)}")
    
    if not is_primitive:
        missing = all_values - unique_values
        print(f"  Fehlende Werte: {sorted(missing)}")
    print()
```

Was ist die Erkenntnis

Von eine wert p muss die basis also g eine prime root sein damit man die gesammt gruppe abdeckt das ist wichtig um beim brutefrocen nicht werte ausblenden kann.
