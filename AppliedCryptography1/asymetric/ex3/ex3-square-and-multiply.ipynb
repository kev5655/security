```python
def square_and_multy(b: int, e: int, n: int) -> int:
    if e == 0:
        return 1
    if e % 2 == 0:
        return square_and_multy(b*b % n, e//2, n)
    else:
        return b * square_and_multy(b, e-1, n) % n
    
print(square_and_multy(3,37,13))
```
