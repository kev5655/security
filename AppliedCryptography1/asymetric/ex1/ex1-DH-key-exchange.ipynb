```python
import os

p = 11
g = 2 # Generator
```

```python
# Alice

# Private key from Alice
a = os.urandom(8) # 8 bytes of random data
print(a.hex())

a = int.from_bytes(a, 'big')
print(a)
```

```python
# Bob

# Private key from Bob
b = int.from_bytes(os.urandom(8), 'big')
```

$$A = g^a \bmod p$$

```python
# Alice generates Public key
A = pow(g, a, p)
```

$$B = g^b \bmod p$$

```python
# Bob generates Public key
B = pow(g, b, p)
```

$$sk = B^a \bmod p = (g^b)^a \bmod p = g^{ba} \bmod p$$

```python
# Alice generates shared secert key

# Resives the Public key B
sk_a = pow(B, a, p)
```

$$sk = A^b \bmod p = (g^a)^b \bmod p = g^{ab} \bmod p$$

```python
# Bob generates shared secret key
sk_b = pow(A, b, p)
```

```python
assert(sk_a == sk_b)
```
