```python
m = b'HiHi'
print(f"Message Bits {len(m)}")

import time
from datetime import datetime, timedelta
from cryptography.hazmat.primitives.asymmetric.ed25519 import Ed25519PrivateKey

private_key = Ed25519PrivateKey.generate()
signature = private_key.sign(m)
public_key = private_key.public_key()

total_iterations = pow(2, len(m) * 8)
print(f"This is the range: {total_iterations}")

start_time = time.time()
print(f"Loop started at: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")

for i in range(total_iterations):
    candidate_bytes = i.to_bytes(len(m), 'big')
    try:
        public_key.verify(signature, candidate_bytes)
        print(f"The message is: {candidate_bytes.decode('utf-8')}")
        break
    except:
        if i % 100000 == 0 and i > 0:
            elapsed = time.time() - start_time
            rate = i / elapsed if elapsed > 0 else 0
            if rate > 0:
                remaining_seconds = (total_iterations - i) / rate
                eta = datetime.now() + timedelta(seconds=remaining_seconds)
                print(
                    f"Now: {datetime.now().strftime('%H:%M:%S')} | "
                    f"Expected finish: {eta.strftime('%H:%M:%S')} | "
                    f"Elapsed: {elapsed:.1f}s | Remaining: {remaining_seconds:.1f}s"
                )
            else:
                print(
                    f"Now: {datetime.now().strftime('%H:%M:%S')} | "
                    "Expected finish: calculating..."
                )
```
