```python
import math

n = 9

def phi(n: int):
    c = 0
    for a in range(1,n+1):
        if math.gcd(a, n) == 1:
            c += 1
    return c
    
# phi(n) = n -1
print(phi(n))

phi_n = phi(n)

for a_i in range(1, n):
    if math.gcd(a_i, n) == 1:
        x = pow(a_i, phi_n, n)  
        print(f"a_i: {a_i} phi_n: {phi_n} n: {n} x: {x}")

        assert x == 1, "Euler Theorem are wrong"
```
