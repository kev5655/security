Bitlänge von 4 sind 0-15 in 0000 - 1111

```python
import math

# How many primes are between 1000 and 9999

# How many primes are between 0-1000
n = 1000
p1 = n / math.log(n)

n = 9999
p2 = n / math.log(n)


estimated_4digit = p2 - p1

exact_under_1000 = 168
exact_under_9999 = 1229
exact_4digit = exact_under_9999 - exact_under_1000

# Vergleich
print(f"Exact: {exact_4digit}")
print(f"Estimate: {estimated_4digit}")
```
