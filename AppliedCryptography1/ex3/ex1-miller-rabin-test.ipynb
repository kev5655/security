```python
# input 4 output range from b'1000' -> 8 to b'1111' -> 15

def getRange(n: int):
    min = pow(2, n-1)
    max = pow(2,n)
    return range(min, max)

r = getRange(4)
assert r[0] == 8, "Min are wrong"
assert r[-1] == 15, "Max are wrong"
```

```python
def search_u_r(n: int) -> tuple[int, int]:
    u = n - 1
    r = 0
    while u % 2 == 0:
        u //= 2
        r += 1
    return u, r
```

```python
import random

# a ist eine zufällige zahl
# n ist die zahl die wir testen wolle ob es ein primzahl ist
def isComposite(a: int, n: int) -> bool:
    u, r = search_u_r(n)
    
    x = pow(a,u,n)
    if x == 1:
        return False
    for _ in range(0,r):
        if x == n-1:
            return False
        x = pow(x,2,n)
    return True

# n ist die zahl die wir testen wolle ob es ein primzahl ist
# t ist die anzahl an versuche umso mehr umso höher die wahrschnlichkeit das das resultat richtig ist.
def miller_rabin(n: int, t: int) -> bool:
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for _ in range(1, t+1):
        a = random.randint(1, n)
        if isComposite(a, n):
            return False
    return True
```

```python
import sympy
bit_range_16 = getRange(16)
t=1

for x in bit_range_16:
    # print(f"Testen ob x={x} eine Prim ist")
    if miller_rabin(x, t):
        if sympy.isprime(x):
            print(f"Found x={x} ist eine prime mit t={t}")
        else:
            print(f"Found Prime but not a prime x={x} mit t={t}")
            
```
