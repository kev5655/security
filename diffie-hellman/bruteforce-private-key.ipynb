# Exercise 3 (Breaking DH Key Agreement)
You are observing that Alice and Bob send each other the values A = 3006 and
B = 2430 in an anonymous DH key agreement with public parameters $\mathbb{Z}^*_{3203}$ and
g = 5. Find the shared secret they agree on.
Hint: Use SageMath which is a Python-like environment with many mathematical
functions.



```python
A = 3006
groupe = 3203  # Z*3203
g = 5
a = 1

while True:
    if (A == pow(g, a, groupe)):
        break
    a += 1

print(f'a is: {a}')
```

```python
B = 2430
b = 1

while True:
    if (A == pow(g, b, groupe)):
        break
    b += 1

print(f'a is: {b}')
```
